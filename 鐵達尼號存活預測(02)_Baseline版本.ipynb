**1.在資料分析的工作流程中，在不考慮準確度情況下有哪些必備的環節？**

(1)資料收集:從原始資料到決定存放資料庫的過程，通常涉及資料獲取、資料爬蟲、資料庫管理、資料倉儲等議題

(2)資料清整:對收集來的資料集做處理，如資料中的錯誤訊息、雜訊等，避免影響模型的準確性

(3)模型訓練:選擇合適的模型(如回歸、分類、機器學習等)進行資料集訓練

(4)決策應用:解釋模型結果，並可透過資料視覺化、圖表的方式呈現上述步驟的結果，同時反思模型的可優化之處

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')
```

```python
# 1.處理缺失值
# 填補年齡的缺失值，用中位數
df['Age'].fillna(df['Age'].median(), inplace=True)
# 填補登船港口的缺失值，用眾數
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
# 填補票價的缺失值，用中位數
df['Fare'].fillna(df['Fare'].median(), inplace=True)

# 使用One-Hot Encoding進行編碼
df_encoded = pd.get_dummies(df, columns=['Sex', 'Embarked'])

# 3.刪除不必要的欄位
# 刪除對預測無用的欄位：PassengerId, Name, Ticket, Cabin
df_train = df_encoded.drop(['PassengerId', 'Name', 'Ticket', 'Cabin'], axis=1)
```

```python
# 準備特徵和目標變數
columns_X = list(set(df_train.columns)-{'Survived'})
columns_y = ['Survived']

train_X = df_train[columns_X]
train_y = df_train[columns_y]

from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# 建立並訓練模型
log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(), cv=5, scoring='accuracy')#交叉驗證

# 輸出交叉驗證分數
print(scores)
```

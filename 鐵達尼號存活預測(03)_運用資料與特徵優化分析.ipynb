```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.impute import SimpleImputer

# 讀取資料
df = pd.read_csv('https://raw.githubusercontent.com/dsindy/kaggle-titanic/master/data/train.csv')

# 1. 資料中哪些欄位是多餘的
# 我們先檢視資料集
print(df.head())
print(df.info())

# 多餘的欄位
# - PassengerId: 是唯一識別符號，對於預測沒有幫助。
# - Name: 雖然有可能提取一些有用的特徵（如稱呼），但在這個任務中可以忽略。
# - Ticket: 無法提取有意義的特徵。
# - Cabin: 缺失值太多，並且提取有用信息較為困難。

df.drop(columns=['PassengerId', 'Name', 'Ticket', 'Cabin'], inplace=True)

# 2. 處理有缺失值的欄位
# - Age: 使用中位數填補
# - Embarked: 使用最多的類別填補

imputer_age = SimpleImputer(strategy='median')
df['Age'] = imputer_age.fit_transform(df[['Age']])

imputer_embarked = SimpleImputer(strategy='most_frequent')
df['Embarked'] = imputer_embarked.fit_transform(df[['Embarked']]).ravel()

print(df.isnull().sum())  # 檢查是否還有缺失值

# 3. 處理非數值型欄位
# - Sex: 使用LabelEncoder
# - Embarked: 使用OneHotEncoder

# LabelEncoder
label_encoder = LabelEncoder()
df['Sex'] = label_encoder.fit_transform(df['Sex'])

# OneHotEncoder
df = pd.get_dummies(df, columns=['Embarked'], drop_first=True)

# 4. 特徵工程手法
# - 創建新變數 FamilySize = SibSp + Parch + 1
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1

# 準備訓練數據
columns_X = set(df.columns) - {'Survived'}
columns_y = ['Survived']

train_X = df[list(columns_X)]
train_y = df[columns_y]

# 訓練模型並進行交叉驗證
log = LogisticRegression(random_state=0, max_iter=3000)
scores = cross_val_score(log, train_X, train_y.values.ravel(), cv=5, scoring='accuracy')
print("Cross-validation accuracy scores:", scores)
print("Mean accuracy:", scores.mean())
print("Standard deviation:", scores.std())
```
